```python
print("Our Quatratic Formula:")
print("2x\u00B2+xy+2y\u00B2=-z\u00B2")
```

```python
import numpy as np
S = np.array([[2,1/2,0],
              [1/2,2,0],
              [0,0,1]])
print("Symmetric Matrix S of Our Quadratic's Coefficients:")
print(S)
```

```python
from numpy import linalg as LA
a, Q = LA.eig(S)
print("Matrix of Eigenvectors:")
print(Q)
```

```python
print("Matrix of Eigenvalues:")
A = (np.diag(a))
print(A)
```

```python
print("Transposed Matrix of Eigenvectors:")
QT = (np.transpose(Q))
print(QT)
```

```python
print("Original Matrix:")
result = np.dot(Q.T,A)
newmatrix = np.dot(result, Q)
print(np.absolute(newmatrix))
print("With the Spectral Theorem, we see that our matrices of" +
      "eigenvalues and eigenvectors reproduce its original matrix.")
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,0.1)
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)
Z = -(2*X**2+X*Y+2*Y**2)**(1/2)


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

plt.title("2x\u00B2+xy+2y\u00B2=-z\u00B2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```
